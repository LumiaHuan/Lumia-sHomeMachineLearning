# Separating data with the maximum margin

The line used to separate the dataset is called a separating hyperplane.

In our simple 2D plots, it’s just a line. But, if we have a dataset with three dimensions, we need a plane to separate the data; and if we have data with 1024 dimensions, we need something with 1023 dimensions to separate the data. What do you call something with 1023 dimensions? How about N-1 dimensions? It’s called a hyperplane. The hyperplane is our decision boundary. Everything on one side belongs to one class, and everything on the other side belongs to a different class.

# Finding the maximum margin

We’d like to find the point closest to the separating hyperplane and make sure this is as far away from the separating line as possible. This is known as margin.

The points closest to the separating hyperplane are known as support vectors.

## Solving small datasets with the simplified SMO

```python
import numpy as np
def loadDataSet(fileName):
    dataMat = []; labelList = []
    with open(fileName) as fr:
        for line in fr.readlines():
            lineArr = line.strip().split('\t')
            dataMat.append([float(lineArr[0]), float(lineArr[1])])
            labelList.append(float(lineArr[2]))
    return dataMat, labelList
```

```python
def selectJrand(i,m):
    j = i
    while (j==i):
        j = int(np.random.uniform(0,m))
    return j
```

```python
def clipAlpha(aj, H, L):
    if aj>H:
        return H
    elif aj<L:
        return L
    else:
        return aj
```

```python
def smoSimple(dataMatIn, classLabels, C, toler, maxIter):
    dataMatrix = np.mat(dataMatIn); labelMat = np.mat(classLabels).transpose()
    b = 0
    m,n = np.shape(dataMatrix)
    alphas = np.mat(np.zeros((m,1)))
    curIter = 0
    while (curIter < maxIter):
        alphaPairsChanged = 0
        for i in xrange(m):
            fXi = float(np.multiply(alphas, labelMat).T*\
                       (dataMatrix*dataMatrix[i,:].T))+b
            Ei = fXi - float(labelMat[i])
            if((labelMat[i]*Ei < -toler) and (alphas[i]<C)) or\
            ((labelMat[i]*Ei > toler) and (alphas[i] > 0)):
                j = selectJrand(i,m)
                fXj = float(np.multiply(alphas, labelMat).T*\
                           (dataMatrix*dataMatrix[j,:].T)) + b
                Ej = fXj - float(labelMat[j])
                alphaIold = alphas[i].copy()
                alphaJold = alphas[j].copy()
                if (labelMat[i] != labelMat[j]):
                    L = max(0, alphas[j] - alphas[i])
                    H = min(C, C + alphas[j] - alphas[i])
                else:
                    L = max(0, alphas[j] + alphas[i] - C)
                    H = min(C, alphas[j] + alphas[i])
                if L==H: print "L==H"; continue
                eta = 2.0 * dataMatrix[i,:]*dataMatrix[j,:].T - \
                dataMatrix[i,:]*dataMatrix[i,:].T - \
                dataMatrix[j,:]*dataMatrix[j,:].T
                if eta >= 0: print "eta>=0"; continue
                alphas[j] -= labelMat[j]*(Ei - Ej)/eta
                alphas[j] = clipAlpha(alphas[j],H,L)
                if (abs(alphas[j] - alphaJold) < 0.00001): print \
                    "j not moving enough"; continue
                alphas[i] += labelMat[j]*labelMat[i]*\
                (alphaJold - alphas[j])
                b1 = b - Ei- labelMat[i]*(alphas[i]-alphaIold)*\
                dataMatrix[i,:]*dataMatrix[i,:].T - \
                labelMat[j]*(alphas[j]-alphaJold)*\
                dataMatrix[i,:]*dataMatrix[j,:].T
                b2 = b - Ej- labelMat[i]*(alphas[i]-alphaIold)*\
                dataMatrix[i,:]*dataMatrix[j,:].T - \
                labelMat[j]*(alphas[j]-alphaJold)*\
                dataMatrix[j,:]*dataMatrix[j,:].T
                if (0 < alphas[i]) and (C > alphas[i]): b = b1
                elif (0 < alphas[j]) and (C > alphas[j]): b = b2
                else: b = (b1 + b2)/2.0
                alphaPairsChanged += 1
                print "iter: {} i:{}, pairs changed {}".format(curIter,i,alphaPairsChanged)
        if (alphaPairsChanged == 0): curIter += 1
        else: curIter = 0
        print 'iteration number: {}'.format(curIter)
    return b, alphas
```

## Speeding up optimization with the full Platt SMO

```python
class optStruct:
    def __init__(self, dataMatIn, classLabels, C, toler, kTup):
        self.X = dataMatIn
        self.labelMat = classLabels
        self.C = C
        self.tol = toler
        self.m = np.shape(dataMatIn)[0]
        self.alphas = np.mat(np.zeros((self.m, 1)))
        self.b = 0
        self.eCache = np.mat(np.zeros((self.m, 2)))
def calcEk(oS, k):
    fXk = float(np.multiply(oS.alphas, oS.labelMat).T*\
                (oS.X*oS.X[k,:].T)) + oS.b
    Ek = fXk - float(oS.labelMat[k])
    return Ek
def selectJ(i, oS, Ei):
    maxK = -1; maxDeltaE = 0; Ej = 0
    oS.eCache[i] = [1,Ei]
    validEcacheList = np.nonzero(oS.eCache[:,0].A)[0]
    if (len(validEcacheList)) > 1:
        for k in validEcacheList:
            if k == i: continue
            Ek = calcEk(oS, k)
            deltaE = abs(Ei - Ek)
            if(deltaE > maxDeltaE):
                maxK = k; maxDeltaE = deltaE; Ej = Ek
        return maxK, Ej
    else:
        j = selectJrand(i, oS.m)
        Ej = calcEk(oS, j)
    return j, Ej
def updateEk(oS,k):
    Ek = calcEk(oS,k)
    oS.eCache[k] = [1, Ek]
    
```

```python
def innerL(i, oS):
    Ei = calcEk(oS,i)
    if ((oS.labelMat[i]*Ei < -oS.tol) and (oS.alphas[i] < oS.C)) or\
    ((oS.labelMat[i]*Ei > oS.tol) and (oS.alphas[i] > 0)):
        j, Ej = selectJ(i, oS, Ei)
        alphaIold = oS.alphas[i].copy();alphaJold = oS.alphas[j].copy()
        if(oS.labelMat[i] != oS.labelMat[j]):
            L = max(0, oS.alphas[j] - oS.alphas[i])
            H = min(oS.C, oS.C + oS.alphas[j] - oS.alphas[i])
        else:
            L = max(0, oS.alphas[j] + oS.alphas[i] - oS.C)
            H = min(oS.C, oS.alphas[j] + oS.alphas[i])
        if L==H: 
            #print "L==H"; 
            return 0
        eta = 2.0 * oS.X[i,:]*oS.X[j,:].T - oS.X[i,:]*oS.X[i,:].T - \
        oS.X[j,:]*oS.X[j,:].T
        if eta >= 0: print "eta>=0"; return 0
        oS.alphas[j] -= oS.labelMat[j]*(Ei - Ej)/eta
        oS.alphas[j] = clipAlpha(oS.alphas[j],H,L)
        updateEk(oS, j)
        if (abs(oS.alphas[j] - alphaJold) < 0.00001):
           # print "j not moving enough";
            return 0
        oS.alphas[i] += oS.labelMat[j]*oS.labelMat[i]*\
        (alphaJold - oS.alphas[j])
        updateEk(oS, i)
        b1 = oS.b - Ei- oS.labelMat[i]*(oS.alphas[i]-alphaIold)*\
        oS.X[i,:]*oS.X[i,:].T - oS.labelMat[j]*\
        (oS.alphas[j]-alphaJold)*oS.X[i,:]*oS.X[j,:].T
        b2 = oS.b - Ej- oS.labelMat[i]*(oS.alphas[i]-alphaIold)*\
        oS.X[i,:]*oS.X[j,:].T - oS.labelMat[j]*\
        (oS.alphas[j]-alphaJold)*oS.X[j,:]*oS.X[j,:].T
        if (0 < oS.alphas[i]) and (oS.C > oS.alphas[i]): oS.b = b1
        elif (0 < oS.alphas[j]) and (oS.C > oS.alphas[j]): oS.b = b2
        else: oS.b = (b1 + b2)/2.0
        return 1
    else: return 0
```

```python
def smoP(dataMatIn, classLabels, C, toler, maxIter, kTup=('lin', 0)):
    oS = optStruct(np.mat(dataMatIn),np.mat(classLabels).transpose(),C,toler, kTup)
    iter = 0
    entireSet = True; alphaPairsChanged = 0
    while (iter < maxIter) and ((alphaPairsChanged > 0) or (entireSet)):
        alphaPairsChanged = 0
        if entireSet:
            for i in range(oS.m):
                alphaPairsChanged += innerL(i,oS)
           # print "fullSet, iter: {} i:{}, pairs changed {}".format(iter,i,alphaPairsChanged)
            iter += 1
        else:
            nonBoundIs = np.nonzero((oS.alphas.A > 0) * (oS.alphas.A < C))[0]
            for i in nonBoundIs:
                alphaPairsChanged += innerL(i,oS)
                # print "non-bound, iter: {} i:{}, pairs changed {}".format(iter,i,alphaPairsChanged)
            iter += 1
        if entireSet: entireSet = False
        elif (alphaPairsChanged == 0): entireSet = True
       # print "iteration number: {}".format(iter)
    return oS.b,oS.alphas
```

```python
def calcWs(alphas,dataArr,classLabels):
    X = np.mat(dataArr); labelMat = np.mat(classLabels).transpose()
    m,n = np.shape(X)
    w = np.zeros((n,1))
    for i in xrange(m):
        w += np.multiply(alphas[i]*labelMat[i],X[i,:].T)
    return w
```

# Using kernels for more complex data

## The radial bias function as a kernel

The radial bias function is a kernel that’s often used with support vector machines.

A radial bias function is a function that takes a vector and outputs a scalar based on the vector’s distance.

```python
def kernelTrans(X, A, kTup):
    m,n = np.shape(X)
    K = np.zeros((m,1))
    if kTup[0]=='rbf':
        for i in xrange(m):
            deltaRow = X[i,:] - A
            K[i] = deltaRow*deltaRow.T
        K = np.exp(-1*K / (2*kTup[1]**2))
    elif kTup[0]=='lin':
        K = X * A.T
    else:
        raise NameError('We dont have {} kernel'.format(kTup[0]))
    return K
```

```python
class optStruct:
    def __init__(self, dataMatIn, classLabels, C, toler, kTup):
        self.X = dataMatIn
        self.labelMat = classLabels
        self.C = C
        self.tol = toler
        self.m = np.shape(dataMatIn)[0]
        self.alphas = np.mat(np.zeros((self.m, 1)))
        self.b = 0
        self.eCache = np.mat(np.zeros((self.m, 2)))
        self.K = np.mat(np.zeros((self.m, self.m)))
        for i in xrange(self.m):
            self.K[:, i] = kernelTrans(self.X, self.X[i,:], kTup)
```

```python
def innerL(i, oS):
    Ei = calcEk(oS,i)
    if ((oS.labelMat[i]*Ei < -oS.tol) and (oS.alphas[i] < oS.C)) or\
    ((oS.labelMat[i]*Ei > oS.tol) and (oS.alphas[i] > 0)):
        j, Ej = selectJ(i, oS, Ei)
        alphaIold = oS.alphas[i].copy();alphaJold = oS.alphas[j].copy()
        if(oS.labelMat[i] != oS.labelMat[j]):
            L = max(0, oS.alphas[j] - oS.alphas[i])
            H = min(oS.C, oS.C + oS.alphas[j] - oS.alphas[i])
        else:
            L = max(0, oS.alphas[j] + oS.alphas[i] - oS.C)
            H = min(oS.C, oS.alphas[j] + oS.alphas[i])
        if L==H: 
           # print "L==H";
            return 0
        eta = 2.0 * oS.K[i,j] - oS.K[i,i] - oS.K[j,j]
        if eta >= 0: print "eta>=0"; return 0
        oS.alphas[j] -= oS.labelMat[j]*(Ei - Ej)/eta
        oS.alphas[j] = clipAlpha(oS.alphas[j],H,L)
        updateEk(oS, j)
        if (abs(oS.alphas[j] - alphaJold) < 0.00001):
            #print "j not moving enough";
            return 0
        oS.alphas[i] += oS.labelMat[j]*oS.labelMat[i]*\
        (alphaJold - oS.alphas[j])
        updateEk(oS, i)
        b1 = oS.b - Ei- oS.labelMat[i]*(oS.alphas[i]-alphaIold)*\
        oS.K[i,i] - oS.labelMat[j]*\
        (oS.alphas[j]-alphaJold)*oS.K[i,j]
        b2 = oS.b - Ej- oS.labelMat[i]*(oS.alphas[i]-alphaIold)*\
        oS.K[i,j] - oS.labelMat[j]*\
        (oS.alphas[j]-alphaJold)*oS.K[j,j]
        if (0 < oS.alphas[i]) and (oS.C > oS.alphas[i]): oS.b = b1
        elif (0 < oS.alphas[j]) and (oS.C > oS.alphas[j]): oS.b = b2
        else: oS.b = (b1 + b2)/2.0
        return 1
    else: return 0
```

```python
def calcEk(oS, k):
    fXk = float(np.multiply(oS.alphas, oS.labelMat).T*\
                (oS.K[:,k])) + oS.b
    Ek = fXk - float(oS.labelMat[k])
    return Ek
```

## Using a kernel for testing

```python
def testRbf(C=200, k1 = 1.5):
    dataArr, labelArr = loadDataSet('testSetRBF.txt')
    b, alphas = smoP(dataArr, labelArr, C, 0.0001, 10000, ('rbf', k1))
    dataMat=np.mat(dataArr);labelVector = np.mat(labelArr).transpose()
    m,n = np.shape(dataMat)
    dataMatPos = dataMat[labelVector.A.flatten()==1]
    dataMatNeg = dataMat[labelVector.A.flatten()==-1]

    fig, ax = plt.subplots(nrows=2, ncols=1, figsize=(8, 14))
    ax[0].scatter(dataMatPos[:,0], dataMatPos[:,1], c='r', s=30)
    ax[0].scatter(dataMatNeg[:,0], dataMatNeg[:,1], c='b', s=30)
    svIndex = np.nonzero((0<alphas.A) * (alphas.A<C))[0]
    supportVectors = dataMat[svIndex]
    supportVectorsLabels = labelVector[svIndex]
    print "k1={} sv.No={}".format(k1, np.shape(supportVectors)[0])
    for index in xrange(len(supportVectors)):
        circle = plt.Circle(tuple(np.array(supportVectors[index]).flatten().tolist()), 0.05, facecolor='none',edgecolor=(0,0.8,0.8), linewidth=3, alpha=0.5)
        ax[0].add_patch(circle)
    errorCount = 0.0
    for i in xrange(m):
        kernelEval = kernelTrans(supportVectors, dataMat[i,:],('rbf',k1))
        predict = kernelEval.T * np.multiply(supportVectorsLabels, alphas[svIndex]) + b
        if np.sign(predict[0,0]) != np.sign(labelArr[i]):
            errorCount += 1.0
    print 'Train error rate:{}'.format(errorCount / float(m))
    dataArrTest, labelArrTest = loadDataSet('testSetRBF2.txt')
    dataMatTest=np.mat(dataArrTest);labelVectorTest = np.mat(labelArrTest).transpose()
    m,n = np.shape(dataMatTest)
    dataMatPosTest = dataMatTest[labelVectorTest.A.flatten()==1]
    dataMatNegTest = dataMatTest[labelVectorTest.A.flatten()==-1]
    
    ax[1].scatter(dataMatPosTest[:,0], dataMatPosTest[:,1], c='r', s=30)
    ax[1].scatter(dataMatNegTest[:,0], dataMatNegTest[:,1], c='b', s=30)
    errorCount = 0.0
    for i in xrange(m):
        kernelEval = kernelTrans(supportVectors, dataMatTest[i,:],('rbf',k1))
        predict = kernelEval.T * np.multiply(supportVectorsLabels, alphas[svIndex]) + b
        if np.sign(predict[0,0]) != np.sign(labelArrTest[i]):
            errorCount += 1.0
            circle = plt.Circle(tuple(np.array(dataMatTest[i]).flatten().tolist()), 0.05, facecolor='none',edgecolor=(1.0,0.0,0.0), linewidth=3, alpha=0.5)
            ax[1].add_patch(circle)
    print 'Test error rate:{}'.format(errorCount / float(m))
```

# Example: revisiting handwriting classification

```python
def img2vector(filename):
    retVector = np.zeros((1,1024))
    with open(filename,'rt') as fr:
        for i in xrange(32):
            line = fr.readline()
            for j in xrange(32):
                retVector[0, i*32+j] = int(line[j])
    return retVector
```

```python
def loadData(fileName):
    from os import listdir
    hwLabels = []
    trainFiles = listdir(fileName)
    m = len(trainFiles)
    trainingMat = np.zeros((m,1024))
    for i in xrange(m):
        label = int(trainFiles[i].split(".")[0].split("_")[0])
        hwLabels.append(label)
        trainingMat[i,:] = img2vector('digits/trainingDigits/'+trainFiles[i])
    return trainingMat, hwLabels
```

```python
def transformHwLabels(recNum, hwLabels):
    retLabels = hwLabels[:]
    for i in xrange(len(hwLabels)):
        if hwLabels[i] == recNum:
            print hwLabels[i]
            retLabels[i] = 1
        else:
            retLabels[i] = -1
    return retLabels
```

```python
def classifyImage(kTup=('rbf',10), recNum=1):
    trainingArr, hwLabelsArr = loadData('digits/trainingDigits/')
    hwLabelsArr = transformHwLabels(recNum, hwLabelsArr)
    b, alphas = smoP(trainingArr, hwLabelsArr, 200, 0.0001, 10000, kTup)
    trainingMat = np.mat(trainingArr); trainingLabelVector = np.mat(hwLabelsArr).transpose()
    svIndex = np.nonzero((0<alphas.A))[0]
    supportVectors = trainingMat[svIndex]
    supportVectorsLabels = trainingLabelVector[svIndex]
    m,n = np.shape(trainingMat)
    errorCount = 0.0
    for i in xrange(m):
        kernelEval = kernelTrans(supportVectors, trainingMat[i,:],kTup)
        predict = kernelEval.T * np.multiply(supportVectorsLabels, alphas[svIndex]) + b
        if np.sign(predict[0,0]) != np.sign(hwLabelsArr[i]):
            errorCount += 1.0
    print 'Train error rate:{}'.format(errorCount / float(m))
    testDataArr, testLabelArr = loadData('digits/testDigits/')
    testLabelArr = transformHwLabels(recNum, testLabelArr)
    testDataMat=np.mat(testDataArr);testLabelVector = np.mat(testLabelArr).transpose()
    m,n = np.shape(testDataMat)
    errorCount = 0.0
    for i in xrange(m):
        kernelEval = kernelTrans(supportVectors, testDataMat[i,:],kTup)
        predict = kernelEval.T * np.multiply(supportVectorsLabels, alphas[svIndex]) + b
        if np.sign(predict[0,0]) != np.sign(testLabelArr[i]):
            errorCount += 1.0
    print 'Test error rate:{:f}'.format(errorCount / float(m))
```
